### [76. Minimum Window Substring](https://leetcode.com/problems/minimum-window-substring)

Given two strings s and t of lengths m and n respectively, return the minimum window substring of s such that every character in t (including duplicates) is included in the window. If there is no such substring, return the empty string "".

The testcases will be generated such that the answer is unique.

 

Example 1:

Input: s = "ADOBECODEBANC", t = "ABC"\
Output: "BANC"\
Explanation: The minimum window substring "BANC" includes 'A', 'B', and 'C' from string t.\
Example 2:

Input: s = "a", t = "a"\
Output: "a"\
Explanation: The entire string s is the minimum window.\
Example 3:

Input: s = "a", t = "aa"\
Output: ""\
Explanation: Both 'a's from t must be included in the window.\
Since the largest window of s only has one 'a', return empty string.
 

Constraints:

- m == s.length
- n == t.length
- 1 <= m, n <= 105
- s and t consist of uppercase and lowercase English letters.

```python
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        len_s = len(s)
        len_t = len(t)

        if len_s < len_t:
            return ""

        res = ""
        # two pointer
        left = 0
        right = 1
        dict_freq_t = {}
        for i in t:
            dict_freq_t[i] = dict_freq_t.get(i, 0) + 1

        # Save current freq substring, which contains character in string t
        dict_cur_match = {}
        # Save current remain freq need to substring incude every character in t
        dict_cur_tmp = dict_freq_t.copy()

        while left < len_s and right <= len_s + 1:
            if right > len_s and dict_cur_tmp:
                break
            # if left > len_s and 
            if not dict_cur_tmp:
                cur_substr_match = s[left: right-1]
                if len(cur_substr_match) < len(res) or res == "":
                    res = cur_substr_match

                if s[left] in dict_freq_t:
                    if dict_cur_match.get(s[left], 0) <= dict_freq_t[s[left]]:
                        dict_cur_tmp[s[left]] = 1
                    dict_cur_match[s[left]] = max(dict_cur_match.get(s[left], 0) - 1, 0)

                left += 1
            else:
                if s[right-1] in dict_cur_tmp:
                    if dict_cur_tmp[s[right-1]] == 1:
                        del dict_cur_tmp[s[right-1]]
                    else:
                        dict_cur_tmp[s[right-1]] -= 1
                if s[right-1] in dict_freq_t:
                    dict_cur_match[s[right-1]] = dict_cur_match.get(s[right-1], 0) + 1
                right += 1

        return res
```
